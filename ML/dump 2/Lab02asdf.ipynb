[Creating multiple subplots](https://matplotlib.org/stable/gallery/subplots_axes_and_figures/subplots_demo.html) -- Pandas docs

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
X = [2, 4, 5, 7, 9]
Y = [5, 4, 10, 2, 8]

fig, axes = plt.subplots(nrows=2, ncols=2)
fig.suptitle('Random charts')

axes[0, 0].set_title('Line plot')
axes[0, 0].plot(X, Y)

axes[0, 1].set_title('Bar chart')
axes[0, 1].bar(X, Y)

axes[1, 0].set_title('Scatter plot')
axes[1, 0].scatter(X, Y)

distribution = np.random.normal(170, 10, 250) 
axes[1, 1].set_title('Histogram')
axes[1, 1].hist(distribution)
```
