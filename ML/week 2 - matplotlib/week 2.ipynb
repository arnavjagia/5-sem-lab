```python
import matplotlib.pyplot as plt
```

```python
# a
fig = plt.figure()
# b
ax = fig.add_axes(rect=(0, 0, 1, 1))
# c
ax.plot([0, 100], [0, 200])
plt.title("title")
plt.xticks(range(0, 101, 20))
plt.yticks(range(0, 201, 50))
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_axes(rect=(0, 0, 1, 1))
ax2 = fig.add_axes(rect=(0.2, 0.5, 0.2, 0.2))
ax1.plot([0, 100], [0, 200])
ax2.plot([0, 100], [0, 200])
plt.show()
```

```python
import pandas as pd
```

```python
df = pd.read_csv("company_sales_data.csv")
month_number = df["month_number"]
total_profit = df["total_profit"]

plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.plot(month_number, total_profit)
plt.show()
```

```python
df = pd.read_csv("company_sales_data.csv")
month_number = df["month_number"]
total_units = df["total_units"]

plt.plot(
    month_number,
    total_units,
    linestyle=':',                   # a. Line Style: dotted
    color='red',                     # a. Line Color: red
    marker='o',                      # e. Marker Style: circle
    markerfacecolor='red',           # f. Marker Face Color: red
    linewidth=3                      # g. Line Width: 3
)

plt.xlabel("Month Number")  # c.
plt.ylabel("Total Units")  # d.
plt.legend(["Total Units"], loc = "lower right")  # b.

plt.show()
```
