```python
import matplotlib.pyplot as plt

X = [0, 100]
Y = [0, 200]
```

```python
fig = plt.figure()
axis = fig.add_axes([0, 0, 1, 1])
axis.plot(X, Y)
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0, 0, 1, 1])
ax1.plot(X, Y)

ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])
ax2.plot(X, Y)
```

```python
import pandas as pd

path_to_data = './data/company_sales_data.csv'

df = pd.read_csv(path_to_data)

month_number = df["month_number"]
total_profit = df["total_profit"]

plt.xlabel("Month")
plt.ylabel("Profit")

plt.plot(month_number, total_profit, 'r:o', linewidth=3)
plt.legend(["Profit"], loc="lower right")

plt.show()
```
